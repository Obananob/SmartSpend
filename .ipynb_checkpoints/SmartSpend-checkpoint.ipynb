```python
import pandas as pd 
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
import seaborn as sns
```

```python
df = pd.read_csv("CC GENERAL.csv")
```

```python
dff = df.dropna().reset_index(drop=True).copy()
```

```python
X = dff.drop(columns = ['CUST_ID'])
### Feature Scaling
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X)
```

```python
# pd.DataFrame(scaled_data).describe()
```

```python
pca = PCA(n_components=5)
reduced_data = pca.fit_transform(scaled_data)
```

```python
pd.DataFrame(reduced_data)
```

## Choose and Configure the Clustering Algorithm

```python
# It's common to start from 2 clusters as silhouette score is not defined for k=1
k_range = range(2, 11) # Test k from 2 to 10

# Initialize lists to store results
silhouette_scores = []

# Iterate through the range of k values
for k in k_range:
    # Initialize and fit KMeans model
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    kmeans.fit(reduced_data)
       
    # Calculate the silhouette score
    score = silhouette_score(reduced_data, kmeans.labels_)
    silhouette_scores.append(score)


# Find the optimal k based on the highest silhouette score
optimal_k_index = silhouette_scores.index(max(silhouette_scores))
optimal_k = k_range[optimal_k_index]
best_silhouette_score = silhouette_scores[optimal_k_index]

print(f"\nOptimal number of clusters (k) based on Silhouette Score: {optimal_k}")
print(f"Highest Silhouette Score achieved: {best_silhouette_score:.4f}")

# Optional: Visualize the silhouette scores
plt.figure(figsize=(8, 6))
plt.plot(k_range, silhouette_scores, marker='o')
plt.title('Silhouette Score vs. Number of Clusters (k)')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()
```

```python
# Choose the Number of Clusters (k)
# This is a crucial step and often determined using methods like the Elbow Method.
n_clusters = optimal_k

# Initialize and Fit the KMeans Model
# n_init='auto' ensures multiple initializations for better results.
kmeans = KMeans(n_clusters=n_clusters, random_state=0, n_init='auto')
kmeans.fit(reduced_data)

# 4. Get Cluster Labels and Centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

print(f"Cluster Labels: {labels}")
print(f"Cluster Centroids: \n{centroids}")

# 5. (Optional) Visualize the Clusters
plt.figure(figsize=(8, 6))
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=labels, cmap='viridis', s=100, alpha=0.8)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()

# # 6. (Optional) Predict Clusters for New Data
# new_data = np.array([[0.5, 1.5], [7, 9]])
# predicted_labels = kmeans.predict(new_data)
# print(f"Predicted labels for new data: {predicted_labels}")
```
